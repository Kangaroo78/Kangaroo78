```python
def os_grid_surrounding_squares(square_to_check) -> list:
  """Returns a list of all available OS grid squares immediately surrounding a given square """

  hundref = square_to_check[0:2]

  inner_easting = int(square_to_check[3:4])
  inner_northing = int(square_to_check[5:6])

  outer_easting = int(square_to_check[2:3])
  outer_northing = int(square_to_check[4:5])

 


  hundref_map = {(0,0):'SV',
  (1,0):'SW',
  (2,0):'SX',
  (3,0):'SY',
  (4,0):'SZ',
  (5,0):'TV',
  (1,1):'SR',
  (2,1):'SS',
  (3,1):'ST',
  (4,1):'SU',
  (5,1):'TQ',
  (6,1):'TR',
  (1,2):'SM',
  (2,2):'SN',
  (3,2):'SO',
  (4,2):'SP',
  (5,2):'TL',
  (6,2):'TM',
  (2,3):'SH',
  (3,3):'SJ',
  (4,3):'SK',
  (5,3):'TF',
  (6,3):'TG',
  (2,4):'SC',
  (3,4):'SD',
  (4,4):'SE',
  (5,4):'TA',
  (1,5):'NW',
  (2,5):'NX',
  (3,5):'NY',
  (4,5):'NZ',
  (1,6):'NR',
  (2,6):'NS',
  (3,6):'NT',
  (4,6):'NU',
  (0,7):'NL',
  (1,7):'NM',
  (2,7):'NN',
  (3,7):'NO',
  (0,8):'NF',
  (1,8):'NG',
  (2,8):'NH',
  (3,8):'NJ',
  (4,8):'NK',
  (0,9):'NA',
  (1,9):'NB',
  (2,9):'NC',
  (3,9):'ND',
  (1,10):'HW',
  (2,10):'HX',
  (3,10):'HY',
  (4,10):'HZ',
  (3,11):'HT',
  (4,11):'HU',
  (4,12):'HP'}

  def is_gridref_legit(search_square):

    found_square = False
    for key, val in hundref_map.items():
            if val == search_square:
                found_square = True

    if found_square:
      ret_val = 'ok'
    else:
      ret_val = 'INVALID GRID REFERENCE'

    return ret_val 
    
  def get_next_letter_code (search_square, easting_change, northing_change ):
    target_square = ''
    def key_from_val(dict, search_val):
      ret_key= None
      for key, val in dict.items():
        if val == search_val:
          ret_key = key
      return ret_key


    coord = key_from_val(hundref_map,search_square)
    if coord:
      x = coord[0]
      y = coord[1]

      new_x= x+easting_change

      new_y= y+northing_change

     

      target_square = hundref_map.get((new_x,new_y))

    return target_square


  def merge_numbers (lc,o_e, i_e, o_n, i_n):
    str_oe = str(o_e)
    str_ie = str(i_e)
    str_on = str(o_n)
    str_in = str(i_n)
    
    if lc == '':
      str_ref = ''
    else:
      str_ref = lc+str_oe+str_ie+str_on+str_in

    return str_ref

  def get_surrounding(letter_code,outer_easting,inner_easting, outer_northing,inner_northing) -> list:

    neighbouring_squares = []

    if is_gridref_legit(letter_code) == 'INVALID GRID REFERENCE':
      neighbouring_squares = [ 'INVALID GRID REFERENCE']
    else:  


    #get left and right neighbours
     
      ret_letter_code = letter_code
      ret_outer_easting  =outer_easting
      ret_inner_easting  =inner_easting
      ret_outer_northing  =outer_northing
      ret_inner_northing  =inner_northing
      
      if inner_easting-1 >= 0:
        ret_inner_easting = inner_easting-1   
      else:
        ret_inner_easting=9
        if outer_easting-1 >= 0:
          ret_outer_easting = outer_easting-1
        else:
          ret_outer_easting=9
          ret_letter_code = get_next_letter_code(letter_code, -1, 0)
      
      ret_square =  merge_numbers (ret_letter_code,ret_outer_easting, ret_inner_easting, ret_outer_northing, ret_inner_northing)
      if ret_square != '':
        neighbouring_squares.append(ret_square)

      ret_letter_code = letter_code
      ret_outer_easting  =outer_easting
      ret_inner_easting  =inner_easting
      ret_outer_northing  =outer_northing
      ret_inner_northing  =inner_northing

      if inner_easting+1 <= 9:
        ret_inner_easting = inner_easting+1   
      else:
        ret_inner_easting=0
        if outer_easting+1 <= 9:
          ret_outer_easting = outer_easting+1
        else:
          ret_outer_easting=0
          ret_letter_code = get_next_letter_code(letter_code, 1, 0)

      ret_square =  merge_numbers (ret_letter_code,ret_outer_easting, ret_inner_easting, ret_outer_northing, ret_inner_northing)
      if ret_square != '':
        neighbouring_squares.append(ret_square)  


      #Get above and below squares

      ret_letter_code = letter_code
      ret_outer_easting  =outer_easting
      ret_inner_easting  =inner_easting
      ret_outer_northing  =outer_northing
      ret_inner_northing  =inner_northing
      
      if inner_northing-1 >= 0:
        ret_inner_northing = inner_northing-1   
      else:
        ret_inner_northing=9
        if outer_northing-1 >= 0:
          ret_outer_northing = outer_northing-1
        else:
          ret_outer_northing=9
          ret_letter_code = get_next_letter_code(letter_code, 0, -1)
      
      ret_square =  merge_numbers (ret_letter_code,ret_outer_easting, ret_inner_easting, ret_outer_northing, ret_inner_northing)
      if ret_square != '':
          neighbouring_squares.append(ret_square)



      ret_letter_code = letter_code
      ret_outer_easting  =outer_easting
      ret_inner_easting  =inner_easting
      ret_outer_northing  =outer_northing
      ret_inner_northing  =inner_northing

      if inner_northing+1 <= 9:
        ret_inner_northing = inner_northing+1   
      else:
        ret_inner_northing=0
        if outer_northing+1 <= 9:
          ret_outer_northing = outer_northing+1
        else:
          ret_outer_northing=0
          ret_letter_code = get_next_letter_code(letter_code, 0, 1)

      ret_square =  merge_numbers (ret_letter_code,ret_outer_easting, ret_inner_easting, ret_outer_northing, ret_inner_northing)
      if ret_square != '':
        neighbouring_squares.append(ret_square)  





    #get the four corners
    #top left
      change_letter_east = False
      change_letter_north = False
      ret_letter_code = letter_code
      ret_outer_easting  =outer_easting
      ret_inner_easting  =inner_easting
      ret_outer_northing  =outer_northing
      ret_inner_northing  =inner_northing

      if inner_northing+1 <= 9:
        ret_inner_northing = inner_northing+1   
      else:
        ret_inner_northing=0
        if outer_northing+1 <= 9:
          ret_outer_northing = outer_northing+1
        else:
          ret_outer_northing=0
          change_letter_north = True

      if inner_easting-1 >= 0:
        ret_inner_easting = inner_easting-1   
      else:
        ret_inner_easting=9
        if outer_easting-1 >= 0:
          ret_outer_easting = outer_easting-1
        else:
          ret_outer_easting=9
          change_letter_east = True
      
      if change_letter_east == True and change_letter_north == False:
        ret_letter_code = get_next_letter_code(letter_code, -1, 0)
      elif  change_letter_east == False and change_letter_north == True:
          ret_letter_code = get_next_letter_code(letter_code, 0, 1)
      elif  change_letter_east == True and change_letter_north == True:
        ret_letter_code = get_next_letter_code(letter_code, -1, 1)

      ret_square =  merge_numbers (ret_letter_code,ret_outer_easting, ret_inner_easting, ret_outer_northing, ret_inner_northing)
      if ret_square != '':
        neighbouring_squares.append(ret_square)


      #top_right = merge_numbers(inner_easting+1, inner_northing+1)
      change_letter_east = False
      change_letter_north = False
      ret_letter_code = letter_code
      ret_outer_easting  =outer_easting
      ret_inner_easting  =inner_easting
      ret_outer_northing  =outer_northing
      ret_inner_northing  =inner_northing

      if inner_northing+1 <= 9:
        ret_inner_northing = inner_northing+1   
      else:
        ret_inner_northing=0
        if outer_northing+1 <= 9:
          ret_outer_northing = outer_northing+1
        else:
          ret_outer_northing=0
          change_letter_north = True
          ret_letter_code = get_next_letter_code(letter_code, 0, 1)

    
      if inner_easting+1 <= 9:
        ret_inner_easting = inner_easting+1   
      else:
        ret_inner_easting=0
        if outer_easting+1 <= 9:
          ret_outer_easting = outer_easting+1
        else:
          ret_outer_easting=0
          change_letter_east = True
          
        
        if change_letter_east == True and change_letter_north == False:
          ret_letter_code = get_next_letter_code(letter_code, 1, 0)
        elif  change_letter_east == False and change_letter_north == True:
          ret_letter_code = get_next_letter_code(letter_code, 0, 1)
        elif  change_letter_east == True and change_letter_north == True:
          ret_letter_code = get_next_letter_code(letter_code, 1, 1)


      ret_square =  merge_numbers (ret_letter_code,ret_outer_easting, ret_inner_easting, ret_outer_northing, ret_inner_northing)
      if ret_square != '':
        neighbouring_squares.append(ret_square)




    # bottom_right = merge_numbers(inner_easting+1, inner_northing-1)
      change_letter_east = False
      change_letter_north = False
      ret_letter_code = letter_code
      ret_outer_easting  =outer_easting
      ret_inner_easting  =inner_easting
      ret_outer_northing  =outer_northing
      ret_inner_northing  =inner_northing

      if inner_northing-1 >= 0:
        ret_inner_northing = inner_northing-1   
      else:
        ret_inner_northing=9
        if outer_northing-1 >= 0:
          ret_outer_northing = outer_northing-1
        else:
          ret_outer_northing=9
          change_letter_north = True

    
      if inner_easting+1 <= 9:
        ret_inner_easting = inner_easting+1   
      else:
        ret_inner_easting=0
        if outer_easting+1 <= 9:
          ret_outer_easting = outer_easting+1
        else:
          ret_outer_easting=0
          change_letter_east = True
          
        
        if change_letter_east == True and change_letter_north == False:
          ret_letter_code = get_next_letter_code(letter_code, 1, 0)
        elif  change_letter_east == False and change_letter_north == True:
          ret_letter_code = get_next_letter_code(letter_code, 0, -1)
        elif  change_letter_east == True and change_letter_north == True:
          ret_letter_code = get_next_letter_code(letter_code, 1, -1)


      ret_square =  merge_numbers (ret_letter_code,ret_outer_easting, ret_inner_easting, ret_outer_northing, ret_inner_northing)
      if ret_square != '':
        neighbouring_squares.append(ret_square)



    # bottom_left = merge_numbers(inner_easting-1, inner_northing-1)
      change_letter_east = False
      change_letter_north = False
      ret_letter_code = letter_code
      ret_outer_easting  =outer_easting
      ret_inner_easting  =inner_easting
      ret_outer_northing  =outer_northing
      ret_inner_northing  =inner_northing

      if inner_northing-1 >= 0:
        ret_inner_northing = inner_northing-1   
      else:
        ret_inner_northing=9
        if outer_northing-1 >= 0:
          ret_outer_northing = outer_northing-1
        else:
          ret_outer_northing=9
          change_letter_north = True

    
    
      if inner_easting-1 >= 0:
        ret_inner_easting = inner_easting-1   
      else:
        ret_inner_easting=9
        if outer_easting-1 >= 0:
          ret_outer_easting = outer_easting-1
        else:
          ret_outer_easting=9
          change_letter_east = True
          
        
        if change_letter_east == True and change_letter_north == False:
          ret_letter_code = get_next_letter_code(letter_code, -1, 0)
        elif  change_letter_east == False and change_letter_north == True:
          ret_letter_code = get_next_letter_code(letter_code, 0, -1)
        elif  change_letter_east == True and change_letter_north == True:
          ret_letter_code = get_next_letter_code(letter_code, -1, -1)


      ret_square =  merge_numbers (ret_letter_code,ret_outer_easting, ret_inner_easting, ret_outer_northing, ret_inner_northing)
      if ret_square != '':
        neighbouring_squares.append(ret_square)
      
      return neighbouring_squares  


  neighbours = get_surrounding(hundref,outer_easting,inner_easting, outer_northing,inner_northing)

  return(neighbours)


```
